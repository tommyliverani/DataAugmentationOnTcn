# Analyze criticalities edges

##### Objectives: 
    1. Obtain anomalies graph
    2. Obtain number of rising edges

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from itables import show, options
options.maxBytes = 0
options.maxColumns = 0

from my_lib import date_utils as du
```

##### 1. Obtain anomalies graph

```python
def plot_timeseries(timestamps, data) :
    # useful for plotting notes
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])

    # corresponding x axis values
    dates = du.extract_dates_from_timestamps(timestamps)
    
    # plotting the points 
    plt.plot(dates, data)
    
    days = []
    i = 0
    dates=dates.unique()
    while i < len(dates):
        days.append(dates[i])
        i+=2
        
    dates = np.asarray(days)

    # naming the x and y axis
    plt.xlabel('days') 
    plt.ylabel('state')
    
    plt.xticks(dates, rotation=35)
    
    # display legend
    plt.text(0.4, 0.9,
             '0: non_rising_anomaly\n1: rising_anomaly',
             bbox={'facecolor': 'pink'}, fontsize=12,
             horizontalalignment='left',
             verticalalignment='center',
             transform = ax.transAxes)
    
    # show the plot 
    plt.show()
```

```python
nodes= ['r089c13s04','r092c08s04','r071c14s04','r169c03s01','r173c16s02','r088c14s01','r095c12s01','r101c09s01','r106c14s01','r076c08s03']
```

```python
for node in nodes:
    print('\n\n\n\n{}\n'.format(node))
    dataset = pd.read_csv('final_data/final_data_{}.csv'.format(node), parse_dates=[0], infer_datetime_format=True, dayfirst=True)
    plot_timeseries(dataset['timestamp'], dataset['is_rising_anomaly'])
    rising_edge_num = len(np.where(np.asarray(dataset['is_rising_anomaly']) == 1)[0])
    print('rising edge:{}'.format(rising_edge_num))
    
```
