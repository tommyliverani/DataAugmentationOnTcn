## Basic TCN using keras-tcn library

* one step forward prediction
* casual network &rarr; `padding` argument of `TCN` layers
* activation function for hidden layers (e.g. 'linear') &rarr; `activation` argument of `TCN` layers
* dropout_rate &rarr; `dropout_rate` argument of `TCN` layers
* skip connections from input to each residual block &rarr; `use_skip_connections` argument of `TCN` layers
* num. stacks of residual blocks &rarr; `nb_stacks` argument of `TCN` layers
* dilation list (e.g. [1, 2, 4, 8, 16] for 32 timesteps receptive field) &rarr; `dilations` arguments of `TCN` layers
* num. convolutional filters (for each convolutional layer) &rarr; `nb_filters` argument of `TCN` layers
* batch normalization &rarr; `use_batch_norm` argument of `TCN` layers


For more details: [keras-tcn](https://github.com/philipperemy/keras-tcn)


```python
import pandas as pd
from itables import show
from itables import options as opt
import numpy as np
import matplotlib.pyplot as plt

opt.maxBytes=0
opt.maxColumns=0

from tensorflow.keras.layers import Dense, Input, Flatten
from tensorflow.keras import Model
from tensorflow.keras.utils import plot_model
from sklearn.metrics import confusion_matrix

from my_lib import tcn_utils as tcnu
from my_lib import general_utils as gu
from my_lib import model_evaluation_utils as meu

from tcn import TCN
```

###### Read Dataset and get training, validation and test data

```python
# Parameters for train, validation and test split
train_test_split_rate = 0.2
validation_split = 0.2
```

```python
dataset, x_train, y_train, x_valid, y_valid, x_test, y_test = gu.data_and_labels_from_csv( \
                                                                        './final_data/final_data_full.csv', \
                                                                        index_cols=['node', 'timestamp'], \
                                                                        labels_cols=[-3, -2, -1], \
                                                                        train_test_split_rate=train_test_split_rate, \
                                                                        train_valid_split_rate=validation_split)
```

```python
dataset
```

```python
# Parameters
n_features = x_train.shape[1]
receptive_field = 128

# TCN
tcn_args = { \
    'dilations': tcnu.get_dilation_rates(receptive_field, 2), \
    'nb_filters': n_features, \
    'kernel_size': 2, \
    'nb_stacks': 1, \
    'dropout_rate': 0.2, \
    'padding': 'causal', \
    'use_skip_connections': True, \
    'return_sequences': True, \
    'activation': 'linear', \
    'kernel_initializer': 'he_normal', \
    'use_batch_norm': True, \
    'name': 'tcn'
}
```

```python
i = Input(shape=(receptive_field, n_features), name='input')
o = TCN(**tcn_args)(i) # The TCN layers are here
o = Flatten(name='flatten')(o)
o = Dense(3, name='output')(o)

model = Model(inputs=[i], outputs=[o])
model.compile(optimizer='adam', loss='categorical_crossentropy')
```

```python
model.summary()
```

```python
plot_model(model, to_file='./models/keras_tcn.png', show_shapes=True, show_layer_names=True)
```

##### Get timeseries generators for training and validation data

```python
# Parameters for generator
train_batch_size = 64
valid_batch_size = 32
```

```python
train_generator = tcnu.get_timeseries_generators(x_train, y_train, receptive_field, train_batch_size)
valid_generator = tcnu.get_timeseries_generators(x_valid, y_valid, receptive_field, valid_batch_size)
```

##### Train the model

```python
# Model fitting parameters
epochs = 10
```

```python
%%time
history = model.fit(train_generator, epochs=epochs, validation_data=valid_generator, \
                    steps_per_epoch=len(train_generator), validation_steps = len(valid_generator))
```

```python
# model.save('./models/keras-tcn_model_10_epochs.h5')
```

##### Visualize gradient trend

```python
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
plt.savefig('./models/keras-tcn_train_loss_plot_10_epochs.png')
```

##### Predict new labels

```python
# Parameters for new labels prediction
test_batch_size = valid_batch_size
```

```python
test_generator = tcnu.get_timeseries_generators(x_test, y_test, receptive_field, test_batch_size)
predicted_probabilities = model.predict(test_generator)
```

```python
# Parameters for label assignment
lb_probability_threshold = 0
ub_probability_threshold = 7.2e-10
```

```python
predicted_rising_anomalies = predicted_probabilities.transpose()[2]
predicted_rising_anomalies_class = meu.assign_class_rising_anomaly(predicted_rising_anomalies, lb_probability_threshold, ub_probability_threshold)
```

#####  CONFUSION MATRIX: Find True Positive (TP), True Negative (TN), False Positive (FP) and False Negative (FN)

```python
y_rising_anomaly_shifted_test = y_test.transpose()[2, receptive_field:]
```

```python
tn, fp, fn, tp = confusion_matrix(y_rising_anomaly_shifted_test, predicted_rising_anomalies_class).ravel()
metrics = meu.compute_metrics(tn, fp, fn, tp)
```

##### Plot the results

```python
%matplotlib notebook

# Build y axes: real rising anomaly -> 1; real non rising anomaly -> 0 
y_val_negative_class = 0
y_val_positive_class = 1
y_vals = [y_val_negative_class if real_rising == 0 else y_val_positive_class \
               for real_rising in y_rising_anomaly_shifted_test]

# Build x axes: rising anomaly prediction value
x_vals = predicted_rising_anomalies

# Set colors: black for predicted non rising; red for predicted rising
colors = ['black' if pred_rising_anomaly_class == 0 else 'red' \
          for pred_rising_anomaly_class in predicted_rising_anomalies_class]

# Build the plot
plt.scatter(x_vals, y_vals, marker='o', s=5, color=colors)

# Set axes limits
axes = plt.gca()
axes.set_xlim([np.min(predicted_rising_anomalies) - 1e-6,np.max(predicted_rising_anomalies) + 1e-6])
axes.set_ylim([y_val_negative_class - 1, y_val_positive_class + 0.5])

# display legend
axes.text(0.4, 0.9,
         'red dot: classified as rising anomaly\nblack dot: classified as non rising anomaly',
         bbox={'facecolor': 'yellow'}, fontsize=8,
         horizontalalignment='left',
         verticalalignment='center',
         transform = axes.transAxes)

# Naming the axes
axes.set_xlabel('rising anomaly prediction value')
axes.set_ylabel('rising anomaly real class')
axes.set_title('Prediction probabilities and real class for rising anomalies')

plt.show()
```
